```python
import pandas as pd
from rdkit.Chem import AllChem, Descriptors
from rdkit import Chem
from rdkit.ML.Descriptors import MoleculeDescriptors
from molvs import Validator, validate_smiles, Standardizer, standardize_smiles
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
```

```python
def generate_descriptors(smiles):    

    molecules = [Chem.MolFromSmiles(smile) for smile in smiles]
    Descriptors._descList
    calc = MoleculeDescriptors.MolecularDescriptorCalculator([desc[0] for desc in Descriptors._descList])
    desc_names = calc.GetDescriptorNames()

    molecule_desc = []
    i = 0
    for molecule in molecules:
        if(i%100==0):
            print("Evaluating ",i)
        descriptors = calc.CalcDescriptors(molecule)
        molecule_desc.append(descriptors)
        i = i + 1
    
    return molecule_desc, desc_names
```

```python
dataset = pd.read_csv('train_II.csv')

dataset[['Chemical ID', 'Assay ID']] = dataset.Id.str.split(';', expand=True)

dataset.drop('Id', axis = 1, inplace=True)

print(dataset)
```

```python
mol_desc, desc_names = generate_descriptors(dataset['Chemical ID'])
```

```python
train_data = pd.DataFrame(mol_desc, columns=desc_names)

train_data.insert(loc=0, column='Assay ID', value=dataset['Assay ID'])

train_data['Expected'] = dataset['Expected']
```

```python
train_data.to_csv('feature_descriptors.csv')
```

```python
test_data = pd.read_csv('test_II.csv')
test_data[['Chemical ID', 'Assay ID']] = test_data.x.str.split(';', expand=True)
test_data.drop('x', axis=1, inplace=True)
```

```python
test_mol_desc, test_desc_names = generate_descriptors(test_data['Chemical ID'])
```

```python
generated_desc = pd.DataFrame(test_mol_desc, columns=test_desc_names)
generated_desc.insert(loc=0, column='Assay ID', value = test_data['Assay ID'])
generated_desc.to_csv('test_feature_desc.csv')
```
